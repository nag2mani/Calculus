```python
from scipy.integrate import quad
import numpy as np
import math
```

```python
# define the function to integrate
def f(x):
  return (np.sin(x)**2 + np.cos(x)**2)**0.5

# integrate the fun from 0 to 2
result, error = quad(f,0,2)
print('The result of the integration is ',result)
print('The error of the integration is ',error)
```

```python
a_list = [i/10 for i in range(20,40) for j in range(10,20)]
x = [i/10 for i in range(10,20)]
b_list = x+x+x+x+x+x+x+x+x+x+x+x+x+x+x+x+x+x+x+x
f_list = []
for a, b in zip(a_list, b_list):
  def f(t):
    return (a**2 * math.sin(t)**2 + b**2 * math.cos(t)**2)**0.5
  result , error = quad(f, 0, 2 * math.pi)
  f_list.append(result)
print(a_list)
print(len(a_list))
print(b_list)
print(len(b_list))
print(f_list)
print(len(f_list))
```

```python
import plotly.express as px
fig = px.scatter_3d(x=a_list, y=b_list, z=f_list)
fig.show()
```
