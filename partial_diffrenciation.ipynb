Differentiation Method :
1.   Numerical
2.   Symbolic
3.   Automatic

```python
#This is for the single variable
import autograd.numpy as np
from autograd import grad
def my_function(x):
  return np.sin(x)

#Find the gradient of my function.
gradient_func = grad(my_function)
gradient = gradient_func(2.0)

print(gradient)
```

```python
def myfunc(X):
  x, y = X
  return x**2+y**2
gradient_func = grad(myfunc)
gradient = gradient_func([3.0,2.0])

print(gradient)
print(type(gradient))
```

```python
#Symbolic Graph of Diffrentiation(M) :
import math
x=[i/10 for i in range(300)]
y_smb=[]
for i in range(len(x)):
  a=math.cos(x[i])
  y_smb.append(a)
```

```python
#automatic Graph of Diffrenciation(M) :
y_auto=[]
for i in range(len(x)):
  gradient_func = grad(my_function)
  gradient = gradient_func(x[i])
  y_auto.append(gradient)
```

```python
#Numerical graph of Diffrenciation(Not matched) :
dx= 0.5 #0.01,0.05,0.5
y_num=[]
for i in range(len(x)):
  a= (math.sin(x[i]+dx) - math.sin(x[i]))/dx
  y_num.append(a)
```

```python
import plotly.graph_objects as go
import plotly.express as px
fig = go.Figure()
fig.add_trace(go.Scatter(name = 'Symbolic', x = x, y = y_smb))

fig.add_trace(go.Scatter(name = 'Numerical', x = x, y = y_num))

fig.add_trace(go.Scatter(name = 'automatic', x = x, y = y_auto))

fig.show()
```
