```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Analytical
lst_t_a = []
lst_x_a = []
t = 0.1
while t<=5:
  lst_t_a.append(t)
  x = math.exp(t)
  lst_x_a.append(x)
  t += 0.1
plt.plot(lst_t_a, lst_x_a, label = 'Analytical Graph of x=e^t')
plt.legend()
```

```python
# Eulers' Method
list_t_euler = []
list_x_euler = []
x_hat = 1
t = 0.0
while t < 5:
  list_t_euler.append(t)
  x_hat = x_hat + x_hat*(0.1)
  list_x_euler.append(x_hat)
  t += 0.1
```

```python
# Combining Both the Graph Analytial and Eular 
plt.plot(lst_t_a, lst_x_a, label= 'Analytical solution')
plt.legend()
plt.plot(list_t_euler, list_x_euler, label = 'Solution By Euler')
plt.legend()
```

```python
# Error between Eular and analytical graph.
error = []
for i in range(len(lst_t_a)):
  error1 = abs((lst_x_a[i] - list_x_euler[i]) / lst_x_a[i])
  error.append(error1)
plt.plot(lst_t_a, error, label = 'Euler Error')
plt.legend()
```

```python
# Modified Euler method (midpoint method) x1 = x0 + dt(x0 + dt*x0/2)
num_mid = [1]
dt = 0.1
for j in range(50):
  x1 = num_mid[j] + dt*(num_mid[j] + dt*num_mid[j]/2)
  num_mid.append(x1)
num_mid.pop()
plt.plot(lst_t_a, num_mid, label="modified Eular")
plt.legend()
```

```python
#measure error between modified Eular method  and analytical method
dt = []
for i in range(50):
  dt.append(i/1000)

# to find error
error = []
for j in range(49):
  res = abs(lst_x_a[j] - num_mid[j])/num_mid[j]
  error.append(res)
dt.pop()
plt.plot(dt, error, label="error between modified Eular method  and analytical method")
plt.legend()
```
