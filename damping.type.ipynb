```python
import plotly.express as px
import math
```

```python
#To Check the type of damping;
def check_damp(A,B,C,x0,v0):
  D=B**2-4*A*C
  if D>0:
    return "Over Damped"
  elif D<0:
      return "Under Damped"
  else:
      return "Critically Damped"
```

```python
def graphtype(A, B, C, x0, v0):
    value = check_damp(A, B, C, x0, v0)
    if value == 'Over Damped':
        print('Damping type is: ', value)
        # calculating d1 and d2 where l1=lambda1 and l2=lambda2
        d1 = (v0 - x0 * l2) / (l1 - l2)
        d2 = (v0 - x0 * l1) / (l2 - l1)
        t_list = [i / 100 for i in range(101)]
        x_list = []
        for t in t_list:
            x_list.append(eval('d1 * math.exp((l1) * t) + d2 * math.exp((l2) * t)'))
    elif value == 'Critically Damped':
        print('Damping type is: ', value)
        t_list = [i / 100 for i in range(1001)]
        x_list = []
        for t in t_list:
            x_list.append(eval('(x0 + (v0 - l1* x0) * t) * math.exp(l1 * t)'))
    else:
        print('Damping type is: ', value)
        t_list = [i / 100 for i in range(1001)]
        x_list = []
        for t in t_list:
            x_list.append(math.exp(alpha * t) * ((x0 * math.cos(beta * t)) + (v0 + alpha * x0 / beta) * math.sin(beta * t)))

    fig = px.line(x = t_list, y = x_list)
    fig.show()
```

```python
# Calculating lamdas and alpha, beta for over damping.
l1 = (-B - (B ** 2 - 4 * A * C) ** 0.5) / (2 * A)
l2 = (-B + (B ** 2 - 4 * A * C) ** 0.5) / (2 * A)
alpha = -B / (2 * A)
beta = math.sqrt(abs((B ** 2 / (4 * A ** 2) - (C / A))))
graphtype(A, B, C, x0, v0)
```
